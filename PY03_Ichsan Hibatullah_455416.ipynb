#### Operasi sederhana

```python
# Program untuk menentukan bilangan ganjil dan genap
n = input("Masukkan Bilangan : ")                       # Variabel yang mendefinisikan fungsi input untuk user memasukkan bilangannya sendiri
n = int(n)                                              # Perubahan tipe data menjadi integer
if (n % 2 == 0):                                        # Pemberian statement if else, syarat bilangan genap yakni sisa hasil bagi dengan bilangan 2 sama dengan 0
    print ("Bilangan tersebut adalah bilangan genap")   # Jika bilangan yang dimasukkan ke kondisi diatas dan menghasilkan nilai yang benar maka akan di print kalimat "Bilangan tersebut adalah bilangan genap", jika salah maka akan lanjut ke statement else
else :                                                  # Jika sisa hasil bagi dengan bilangan 2 tidak sama dengan 0 maka bilangan ganjil
    print ("Bilangan tersebut adalah bilangan ganjil")  # Jika bilangan yang dimasukkan ke kondisi diatas dan menghasilkan nilai yang salah maka akan di print kalimat "Bilangan tersebut adalah bilangan ganjil"
```

#### Menggunakan Library Numpy 

```python
import numpy as np                       # Import modul numpy dalam hal ini untuk menyelesaikan persamaan sinus dan cosinus, menggunakan variabel radians (mendeklarasikan pi)
import matplotlib.pyplot as plt          # Import modul matplotlib untuk memvisualisasikan dalam bentuk grafik

x = np.arange (0, 2 * np.pi, np.pi/100)  # Batasnya, rumus dari sinus
y = np.sin (x)                           # Balik dari rumus di atas, nilai sin x dari np.arange di atas
plt.plot (x,y)                           # Untuk plot nilai dari x dan y
plt.grid()                               # Memberi grid (garis-garis/kotak) di belakang
plt.tick_params (labelcolor='tab:green') # Memberi label dan mengganti warna label
plt.show()                               # Untuk menampilkan grafik
```

```python
import numpy as np                                   # Import modul numpy dalam hal ini untuk menyelesaikan persamaan sinus dan cosinus, menggunakan variabel radians (mendeklarasikan pi)
import matplotlib.pyplot as plt                      # Import modul matplotlib untuk memvisualisasikan dalam bentuk grafik

x = np.arange (0, 2 * np.pi, np.pi/100)              # Batasnya, rumus dari sinus
y1 = np.sin (x)                                      # Parameter pertama, nilainya balik dari rumus di atas, nilai sin x dari np.arange di atas
y2 = np.sin (x + np.pi)                              # Parameter kedua, nilainya balik dari rumus di atas, nilai sin x + pi dari np.arange di atas dengan sedikit variasi ditambah dengan pi
plt.plot (x,y1,'*r', label = 'sin(x)')               # Untuk plot nilai dari x dan y1, memodifikasi garisnya, ditambah nama di legenda 
plt.plot (x,y2,'+b', label = 'sin (x+pi)')           # Untuk plot nilai dari x dan y2, memodifikasi garisnya, ditambah nama di legenda 
plt.xlabel ('X')                                     # Pemberian label di sumbu x
plt.ylabel ('Y')                                     # Pemberian label di sumbu y
plt.title ('Plot Data Sinus')                        # Menuliskan Judul
plt.grid()                                           # Memberikan grid (garis-garis/kotak) dibelakang
plt.legend()                                         # Menampilkan legenda yang dari plot grafik di atas
plt.tick_params (axis='both', labelcolor='tab:blue') # Untuk mengganti warna axisnya (X dan Y)
plt.show()                                           # Menampilkan hasil
```

#### Menggunakan Piechart

```python
import matplotlib.pyplot as plt                     # Import modul matplotlib untuk memvisualisasikan dalam bentuk pie chart
labels = 'motor', 'mobil', 'sepeda'                 # Nilai# Memberi judul variabel "labels" dengan data kendaraan
Nilai = [30, 40, 40]                                # Nilai dari masing-masing variabel yang digunakan
fig1, ax1 = plt.subplots()                          
ax1.pie(Nilai, labels=labels, autopct = '%1.1f%%', startangle=90)
ax1.axis ('equal')                                  
ax1.set_title("Kendaraan Pribadi")                  # Memberi judul
plt.show()                                          # Menampilkan grafik
```

##### Stasiun Pengamatan 1 Davao, Philipphines (DAV)

```python
import matplotlib.pyplot as plt                                            # Import modul matplotlib untuk memvisualisasikan gambar grafik gempa bumi
from obspy import read                                                     # Import obspy untuk membaca data gelombang seismik pada data gempa bumi

komp1 = read ("dav.miniseed")                                              # Membaca data gempa bumi di Mamuju dari stasiun DAV (Davao, Philippines)
data1 = komp1 [0].data 
t     = range (1, len(data1)+1) 
fig   = plt.figure()                                                       # Untuk plot gambar grafik
ptl   = plt.subplot(311)                                                   # Menggunakan 311 agar amplitudo data terbaca
ptl.plot(t,data1)                                                          # Plot data yang akan muncul pada grafik berupa gelombang hasil pembacaan serta mengganti warna
ptl.set_xlim([0, len(data1)+1])                                            # Fungsi xlim dibangun di atas fungsi lain, xlim membatasi nilai longitude minimal (terdekat) ke nilai longitude maksimal (terjauh)
plt.ylabel('Amplitudo')                                                    # Memplot grafik amplitudo dengan label berupa besarnya amplitudo pada sumbu y
fig.suptitle('Plot Data Seismik Gempa Mamuju - DAV (Davao, Philipphines)') # Memberikan judul pada bagian header grafik gelombang gempa
plt.xlabel('Time')                                                         # Memplot label time pada sumbu x dan mengganti warna
plt.grid()                                                                 # Memberikan grid (garis-garis/kotak) di belakang
plt.tick_params()                                                          # Untuk mengganti warna axisnya (X dan Y)
plt.show()                                                                 # Menampilkan grafik
```
