TUGAS I

Buatlah suatu program sederhana untuk menyelesaikan :

<ol>
<li>Mencari Kontur Interval (CI) dengan memberikan masukan oleh user berupa penyebut skala </li>
<li>Mencari mencari nilai ketinggian(h) dan kecepatan(v) setelah bola dilemparkan dengan h0 = 8 m dan v0 = 2 m/s dengan variasi t yakni 0.5 s, lebih dari 0.5 s, dan lebih dari 2s. Besar g =9.8 </li>
</ol>

$$h = h_0 + V_0t-\frac12 gt^2$$

$$v = v_0 - gt$$

**Script code :**

No. 1 <br>

```python
print ("\033[1m" + "\n\nProgram Sederhana Mencari Kontur Interval / Beda Ketinggian (CI)" + "\033[0m") #judul program
print (" ") #spasi

skala = input("Masukkan skala (tidak menggunakan titik) (1:x), x = ") #mendefinisikan variabel skala yang berisi fungsi input untuk memasukkan nilai skala dengan tipe data string

print("Skalanya adalah 1:" + skala) #menuliskan kembali perbandingan skala yang akan dimasukkan dalam perhitungan
print (" ")

kontur = 1 / 2000 * float(skala) #mendefinisikan variabel kontur yang berisi perhitungan kontur interval, variabel skala diubah ke dalam tipe data float

print ("Kontur Interval atau Beda Ketinggian yang di dapat adalah " + str(kontur) + " meter") #menuliskan dalam output hasil dari perhitungan yang telah dilakukan, variabel kontur diubah ke dalam tipe data string
```

No. 2 <br>

```python
print("\033[1m" + "Program Sederhana Mencari Nilai Ketinggian (h) dan Kecepatan (v) dengan Variasi Waktu"  + "\033[0m") #judul program
print(" ") #spasi

#Mendefinisikan variabel - variabel yang akan digunakan dalam perhitungan
h0 = 8     #variabel h0 (ketinggian awal)
v0 = 2     #variabel v0 (kecepatan awal)
g  = 9.8   #variabel g  (gravitasi)

#ketika t = 0.5 s (variasi pertama)
t = 0.5    #variabel t  (waktu)
h = (h0+v0*t)-(0.5*g*t**2) #rumus perhitungan ketinggian
v = v0-(g*t) #rumus perhitungan kecepatan
print("Ketinggian benda ketika t=" + str(t) + " adalah " + "%.2f" %h) #menuliskan dalam output hasil dari perhitungan yang telah dilakukan
print(" Kecepatan benda ketika t=" + str(t) + " adalah " + "%.2f" %v)
print (" ")

#ketika t = 1 s (variasi kedua)
t = 1      #variabel t  (waktu)
h = (h0+v0*t)-(0.5*g*t**2) #rumus perhitungan ketinggian
v = v0-(g*t) #rumus perhitungan kecepatan
print("Ketinggian benda ketika t=" + str(t) + " adalah " + "%.2f" %h) #menuliskan dalam output hasil dari perhitungan yang telah dilakukan
print(" Kecepatan benda ketika t=" + str(t) + " adalah " + "%.2f" %v)
print (" ")

#ketika t = 5 s (variasi ketiga)
t = 5      #variabel t  (waktu)
h = (h0+v0*t)-(0.5*g*t**2) #rumus perhitungan ketinggian
v = v0-(g*t) #rumus perhitungan kecepatan
print("Ketinggian benda ketika t=" + str(t) + " adalah " + "%.2f" %h) #menuliskan dalam output hasil dari perhitungan yang telah dilakukan
print(" Kecepatan benda ketika t=" + str(t) + " adalah " + "%.2f" %v)
```

 1. Membuat plot sederhana lokasi gempa berdasarkan data Gempa BMKG dari data excel yang telah dibagikan

**Script code:**

```python
from pandas import read_excel #untuk membaca data yang terdapat di excel #bisa ditambahin lagi pakai as
import cartopy.crs as ccrs #untuk memproyeksikan koordinat baru dengan (PlateCarree)
import matplotlib.pyplot as plt #keseluruhan script menggunakan modul ini sehingga sangat penting
from matplotlib.image import imread #untuk membaca gambar peta

Data = read_excel('Data Gempa BMKG.xlsx') #Mendefinisikan variabel Data untuk membaca data yang terdapat pada excel

plt.figure(figsize=(10,9)) #Membuat kanvas dari sebuah fungsi, diisikan sebuah fungsi atau parameter figsize dengan ukuran 10 x 9
plt.axes(projection=ccrs.PlateCarree()) #Membuat koordinat baru dengan **proyeksi** ccrs.PlateCarree, mengatur sumbu, menggunakan fungsi axes yang diambil dari matplotlib
plt.imshow(imread('Gambar.TIF'), extent=[-180, 180, -90, 90]) #Gambar hasil citra satelit, gambar TIF, Python tidak bisa membaca gambar dari citra satelit, sehingga menggunakan fungsi imread yang di ambil dari matplotlib.image dan juga menggunakan argumen extent, extent = membatasi dimana letak Indonesia (ngezoom ke wilayah Indonesia aja), karena data - datanya hanya di Indonesia 
plt.xlim([min(Data['Longitude']),max(Data['Longitude'])]) #Fungsi xlim dibangun diatas fungsi lain, xlim membatasi nilai Longitude minimal (terdekat) ke nilai Longitude maksimal (terjauh)
plt.ylim([min(Data['Latitude']),max(Data['Latitude'])]) #Fungsi ylim dibangun diatas fungsi lain, ylim membatasi nilai Latitude minimal (terdekat) ke nilai Latitude maksimal (terjauh)
plt.scatter(Data['Longitude'],Data['Latitude'],c='black') #Memasukkan dan menitikkan lokasi - lokasi gempa di wilayah Indonesia dari data - data gempa (longitude dan latitude) yang telah disediakan, titiknya disini diberi warna hitam
plt.title('PERSEBARAN GEMPA 2020') #Menuliskan Judul
plt.show() #Menampilkan Gambar
```
